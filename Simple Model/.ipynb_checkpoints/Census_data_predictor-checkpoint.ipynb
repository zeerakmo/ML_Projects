```python
census_data = census_data.drop(columns=['fnlwgt',
                                       'education-num'], axis=1)
census_data.head()
```

```python
census_data = census_data.dropna()
census_data
```

```python
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()

census_data['class'] = label_encoding.fit_transform(census_data['class'].astype(str))
census_data.sample(10)
```

```python
census_data['workclass'] = label_encoding.fit_transform(census_data['workclass'].astype(str))
census_data['education'] = label_encoding.fit_transform(census_data['education'].astype(str))
census_data['marital-status'] = label_encoding.fit_transform(census_data['marital-status'].astype(str))
census_data['occupation'] = label_encoding.fit_transform(census_data['occupation'].astype(str))
census_data['relationship'] = label_encoding.fit_transform(census_data['relationship'].astype(str))
census_data['race'] = label_encoding.fit_transform(census_data['race'].astype(str))
census_data['sex'] = label_encoding.fit_transform(census_data['sex'].astype(str))
census_data['native-country'] = label_encoding.fit_transform(census_data['native-country'].astype(str))
```

```python
X = census_data.drop('class',axis = 1)
Y = census_data['class']
```

```python
from sklearn import preprocessing

standard_scaler = preprocessing.StandardScaler()
```

```python
features_scaled = standard_scaler.fit_transform(X)
features_scaled.shape
```

```python
features_scaled_df = pd.DataFrame(features_scaled,columns = X.columns)
features_scaled_df.head()
```

```python
census_data = pd.concat([features_scaled_df, census_data['class']],axis=1).reset_index(drop=True)

census_data.head(10)
```

```python
X = census_data.drop('class',axis=1)
Y=census_data['class']
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state= 0)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(penalty = 'l2',C=1.0,solver = 'liblinear')

classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
model_accuracy = accuracy_score(y_test,y_pred)
mode_precision = precision_score(y_test,y_pred)
model_recall = recall_score(y_test,y_pred)

#how many were correct is accuracy
#how many of this models positive results were correct prescision
#how many of the indi with diabeetes across dataset were corrrectly calssified by model is recall

print(model_accuracy)
print(mode_precision)
print(model_recall)
```
