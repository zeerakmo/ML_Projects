```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
diabetes_data = pd.read_csv('Dataset/PimaIndians.csv')
diabetes_data.head()
```

```python
diabetes_data.describe(include='all')
```

```python
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()

diabetes_data['test'] = label_encoding.fit_transform(diabetes_data['test'].astype(str))
diabetes_data.sample(10)
```

```python
diabetes_data = diabetes_data.dropna()
```

```python
diabetes_data.shape
```

```python
X = diabetes_data.drop('test',axis = 1)
Y = diabetes_data['test']
```

```python
X.head()
```

```python
from sklearn import preprocessing

standard_scaler = preprocessing.StandardScaler()
```

```python
features_scaled = standard_scaler.fit_transform(X)
features_scaled.shape
```

```python
features_scaled_df = pd.DataFrame(features_scaled,columns = X.columns)
features_scaled_df.head()
```

```python
features_scaled_df.describe()
```

```python
diabetes_data = pd.concat([features_scaled_df, diabetes_data['test']],axis=1).reset_index(drop=True)

diabetes_data.head()
```

```python
diabetes_data.to_csv('Dataset/PimaIndians_processed.csv',index=False)
```

```python
 !ls Dataset
```

```python
X = diabetes_data.drop('test',axis=1)
Y=diabetes_data['test']
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state= 0)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(penalty = 'l2',C=1.0,solver = 'liblinear')

classifier.fit(x_train,y_train)
```

```python
classifier.score(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
pred_results = pd.DataFrame({'y_test':y_test,
                            'y_pred': y_pred})
pred_results.head(10)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

model_accuracy = accuracy_score(y_test,y_pred)
mode_precision = precision_score(y_test,y_pred)
model_recall = recall_score(y_test,y_pred)

#how many were correct is accuracy
#how many of this models positive results were correct prescision
#how many of the indi with diabeetes across dataset were corrrectly calssified by model is recall

print(model_accuracy*100)
print(mode_precision*100)
print(model_recall*100)
```
