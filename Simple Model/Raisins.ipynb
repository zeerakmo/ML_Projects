```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

```python
raisin_data = pd.read_csv('Dataset/Raisin_Dataset/Raisin_Dataset.csv')
raisin_data.head()
```

```python
from sklearn import preprocessing

label_encoding = preprocessing.LabelEncoder()

raisin_data['Class'] = label_encoding.fit_transform(raisin_data['Class'].astype(str))


```

```python
raisin_data.head()
```

```python
raisin_data = raisin_data.dropna()
```

```python
raisin_data.corr()
```

```python
X = raisin_data.drop('Class',axis=1)
Y = raisin_data['Class']
```

```python
from sklearn import preprocessing

standard_scaler = preprocessing.StandardScaler()

features_scaled = standard_scaler.fit_transform(X)
features_scaled.shape

features_scaled_df = pd.DataFrame(features_scaled, columns = X.columns)
features_scaled_df.head(10)

raisin_data = pd.concat([features_scaled_df,raisin_data['Class']],axis=1).reset_index(drop = True)
```

```python
raisin_data.head()
```

```python
raisin_data_corr= raisin_data.corr()
```

```python
import seaborn as sns

fig,ax = plt.subplots(figsize = (8,8))
sns.heatmap(raisin_data_corr,annot=True)
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

n_estimators = [100,150,200,250,300,500]
learn_rates = [0.001,0.01, 0.02, 0.05, 0.1,0.2,0.3]

param_grid = {'n_estimators': n_estimators,
              'learning_rate': learn_rates}

grid_search = GridSearchCV(GradientBoostingClassifier(random_state = 0),param_grid, cv=2, return_train_score=True,verbose = 10)
grid_search.fit(x_train, y_train)

grid_search.best_params_
```

```python
gradient_boosting = GradientBoostingClassifier(learning_rate=grid_search.best_params_['learning_rate'],n_estimators = grid_search.best_params_['n_estimators'])
gradient_boosting.fit(x_train, y_train)
y_pred = gradient_boosting.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score
model_accuracy = accuracy_score(y_test,y_pred)
mode_precision = precision_score(y_test,y_pred)
model_recall  = recall_score(y_test,y_pred)

#how many were correct is accuracy
#how many of this models positive results were correct prescision
#how many of the indi with diabeetes across dataset were corrrectly calssified by model is recall

print(model_accuracy*100)
print(mode_precision*100)
print(model_recall*100)
```
